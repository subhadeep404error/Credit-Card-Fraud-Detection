```python
#separating the data for analysis
legit=credit_card[credit_card.Class == 0]
fraud=credit_card[credit_card.Class == 1]
```

```python
# Under-Sampling
# Build a sample dataset containing similar distribution of normal transaction and frad transaction
```

```python
legit_sample=legit.sample(n=492)
```

```python
#join two dataframes 

new_data=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data.sample(5)
```

```python
new_data['Class'].value_counts() # the new_data is perfectly balanced 
```

```python
new_data.groupby('Class').mean()
```

```python
# Splitting the data into Features & Targets

X=new_data.drop("Class",axis=1)
Y=new_data['Class']
```

```python
print(X)
print(Y)
```

```python
# Spliting the data into Training data and Testing data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# Model Training
# Logistic Regression

model= LogisticRegression()
```

```python
# Training the logistic regression model with training data
model.fit(X_train, Y_train)
```

```python
# Model Evaluation

# accuracy score

X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training Data: ',training_data_accuracy)
```

```python
# Accuracy on X_test data
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy score on Test data: ' ,testing_data_accuracy)
```
